# Some problems on plotting
> **Problem**: Plot $f(x) = \frac {sin x}{x}$ in the range [-10,10] in python.
## Think to draw
As we have $sin(x)$ it might be sinosoidal. But there have $\frac{1}{x}$ who determine the amplitude. $\frac{1}{x}$ acts as the envelope of the $sin (x)$.
## Help:
You might use `linspace()` or `arange()` from **numpy** to genarate the array. But computer cannot genarate the value of the function at $x=0$. Here we going to remove this discontinuity. If we use limit from both side it tells us that if $f(x)$ must be 1 as $x=0$. 

Where numpy helps us to define the function in a new way using `where`.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

def f(x):
       return np.where(x == 0, 1, np.sin(x) / x)

plt.plot(x, f(x), label=r"$\frac{\sin(x)}{x}$", color="crimson")
plt.grid(linestyle="dotted")
plt.xlabel("$x$")
plt.ylabel(r"$\frac{\sin(x)}{x}$")
plt.legend()
plt.show()
```
